```python
import seaborn as sb
plt.figure(figsize=(10,5),dpi=280)
sb.FacetGrid(data,hue="make",size=10).map(plt.scatter,"width","height").add_legend()
plt.title("HEIGHT VS WIDTH")
plt.show()
```

```python
figure=plt.figure(figsize=(24,14),dpi=280)
sb.boxplot(hue="make",x="height",y="width",data=data)
plt.title("HEIGHT VS WIDTH")
plt.show()
```

```python
plt.figure(figsize=(20,4),dpi=280)
sb.stripplot(x="stroke",y="horsepower",data=data,jitter=True,edgecolor='Cyan')
plt.xlabel("CARS STROKE")
plt.ylabel("CARS HORSEPOWER")
plt.title("HORSEPOWER VS STROKE")
plt.show()
```

```python
figure=plt.figure(figsize=(26,5),dpi=380)
sb.violinplot(x="make",y="price",data=data,size=20)
plt.xlabel("MAKE(CARS BRAND)")
plt.ylabel("PRICE OF THE CAR")
plt.title("Violin plot for Price vs Make")
plt.show()
```

```python
plt.figure(figsize=(10,4),dpi=280)
sb.FacetGrid(data,hue="fuel-type",size=6).map(sb.kdeplot,"price",).add_legend()
plt.xlabel('PRICE OF THE CARS')
plt.ylabel('PROBABILITY DENSITY FUNCTION')
plt.title('Kernel density estimation')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
plt.figure(figsize=(10,4),dpi=280)
sns.set(style='whitegrid',context='notebook')
cols=['normalized-losses', 'make','fuel-type', 'aspiration', 'num-of-doors',
       'body-style', 'drive-wheels', 'engine-location', 'wheel-base', 'length',
       'width', 'height', 'curb-weight', 'engine-type', 'num-of-cylinders',
       'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-ratio',
       'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
sns.pairplot(data[cols],height=4.5)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(8,3),dpi=280)
cm = np.corrcoef(data[cols].values.T)

sns.set(font_scale=1)
plt.title("CORRELATION MATRIX")
hm = sns.heatmap(cm,
                cbar=True,
                annot=True,
                square=True,
                fmt='1.1f',
                annot_kws={'size':8},
                yticklabels=cols,
                xticklabels=cols)
plt.show()
```
